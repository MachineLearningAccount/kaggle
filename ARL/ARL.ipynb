## Выполнение работы:

Импортируем все необходимые библиотеки для работы. Будем искать ассоциативные правила методом apriory.

```python
import pandas as pd
import shutil, os
import apyori as ay
import json
```

Импортируем датасет из следующего источника:

https://www.kaggle.com/mathchi/online-retail-data-set-from-ml-repository

```python
mbasket_df = pd.read_csv('retail_dataset.csv', header=0, sep=',')
display(mbasket_df.head())
print(mbasket_df.info())
```

Датасет представляет собой объекты-транзакции, каждый из которых содержит набор товаров, которые были куплены. Они распологаются в левых колонках по товару на колонку. Колонок 7 и если товаров в транзакции меньше, то оставшиеся колонки справа заполняются NaN. Всего имеем 315 транзакций.

Как можем видеть, все 315 значений не пусты только для товара с индексом 0, а для остальных, есть пустые значения. Заменим NaN на последнее значение внутри транзакции.

```python
mbasket_df.fillna(method = 'ffill',axis = 1, inplace = True)
print(mbasket_df.info())
display(mbasket_df.head(20))
```

Как видно, у нас больше нет пустых ячеек, а заполнились они теперь последним товаром для каждого из объектов.

Теперь, сделаем из нашего df матрицу и сохраним её в переменную transaction.

```python
transactions = []
for i in range(0, 314): 
    transactions.append([str(mbasket_df.values[i,j]) for j in range(0, 6)])
```

```python
print(*transactions, sep="\n")
```

Выполним поиск ассоциативных правил. со следующими требованиями. 

Поддержку сделаем минимум 0.1.

Достоверность сделаем минимум 0.6.

Подъём сделаем минимум 1.1.

Максимальнуд длину itemset сделаем 3.

```python
result = list(ay.apriori(transactions, min_support = 0.1, min_confidence = 0.6, min_lift = 1.1, max_length = 3))
```

Определим количество найденных правил.

```python
len(result)
```

Выведем на экран полный список найденных правил, применив удобный механизм для отображения.

```python
try:
    from StringIO import StringIO
except ImportError:
    from io import StringIO

output = []
for RelationRecord in result:
    o = StringIO()
    ay.dump_as_json(RelationRecord, o)
    output.append(json.loads(o.getvalue()))
data_df = pd.DataFrame(output)

pd.set_option('display.max_colwidth', None)

from IPython.display import display, HTML

display(HTML(data_df.to_html()))
```

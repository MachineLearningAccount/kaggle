Импортируем выборку и посмотрим количество рядов.

```python
tuse_df = pd.read_csv('telecom_users.csv', header = 0, sep = ',')
len(tuse_df)
```

Выведем несколько строк выборки, чтобы визуально взглянуть на данные. Посмотрим информацию по выборке штатными средствами pandas, уточним, есть ли пустые значения. Как видно, пустых значений нет. Количество выборки 5986 строк (абонентов).

```python
display(tuse_df.head(10))
display(tuse_df.info())
```

Как видно по смыслу, первые 2 столбца не имеют никакого значения для нашего анализа, поэтому, не будем их использовать и сократим выборку, убрав их.

```python
tuse_df = pd.DataFrame(tuse_df.iloc[:, 2:])
```

Для корректной работы метода k-means из библиотеки sklearn, выразим категориальные и бинарные переменные в целочисленном представлении для полей:

gender

Partner

Dependents

PhoneService

MultipleLines

InternetService

OnlineSecurity

OnlineBackup

DeviceProtection

TechSupport

StreamingTV

StreamingMovies

Contract

PaperlessBilling

PaymentMethod

Churn

```python
tuse_df['gender'] = pd.factorize(tuse_df['gender'])[0]
tuse_df['Partner'] = pd.factorize(tuse_df['Partner'])[0]
tuse_df['Dependents'] = pd.factorize(tuse_df['Dependents'])[0]
tuse_df['PhoneService'] = pd.factorize(tuse_df['PhoneService'])[0]
tuse_df['MultipleLines'] = pd.factorize(tuse_df['MultipleLines'])[0]
tuse_df['InternetService'] = pd.factorize(tuse_df['InternetService'])[0]
tuse_df['OnlineSecurity'] = pd.factorize(tuse_df['OnlineSecurity'])[0]
tuse_df['OnlineBackup'] = pd.factorize(tuse_df['OnlineBackup'])[0]
tuse_df['DeviceProtection'] = pd.factorize(tuse_df['DeviceProtection'])[0]
tuse_df['TechSupport'] = pd.factorize(tuse_df['TechSupport'])[0]
tuse_df['StreamingTV'] = pd.factorize(tuse_df['StreamingTV'])[0]
tuse_df['StreamingMovies'] = pd.factorize(tuse_df['StreamingMovies'])[0]
tuse_df['Contract'] = pd.factorize(tuse_df['Contract'])[0]
tuse_df['PaperlessBilling'] = pd.factorize(tuse_df['PaperlessBilling'])[0]
tuse_df['PaymentMethod'] = pd.factorize(tuse_df['PaymentMethod'])[0]
tuse_df['Churn'] = pd.factorize(tuse_df['Churn'])[0]

tuse_df.info()
display(tuse_df)
```

Поле TotalCharges осталось объектным. нужно его преобразовать к вещественному типу, так как это валюта. Но, некоторые значения там не заполнены. Вопрос, какие. Как удалось выяснить, не заполнены значения для тех абонентов, для которых время их "клиентства" в компании равно 0. То есть, им платить было не за что и их плата равна 0.

```python
tuse_df[tuse_df["TotalCharges"]==' '].index == tuse_df[tuse_df["tenure"]==0].index

tuse_df["TotalCharges"] = tuse_df["TotalCharges"].replace('[\s+,]', '0', regex=True)
```

```python
tuse_df["TotalCharges"] = tuse_df["TotalCharges"].astype(float)
tuse_df.info()
```

```python
tuse_df[tuse_df['tenure']==0]
```

Теперь попытаемся найти оптимальное количество кластеров методом "Локтя". Попробуем построить график, для интервала секторов от 1 до 10.

```python
wcss = []
for i in range(1, 11):
    kmens = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmens.fit(tuse_df)
    wcss.append(kmens.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

Судя по графику, наиболее оптимальное количество кластеров, это 2 или 3.

Применим метод k-means для количества кластеров = 3.

```python
model = KMeans(n_clusters=3)

model.fit(tuse_df)

tuse_df['cluster'] = model.fit_predict(tuse_df)
tuse_df.tail(15)
```

```python
scatter_fig = plt.figure()
axes = scatter_fig.add_axes([0,0,2,2])
axes.scatter(tuse_df["MonthlyCharges"], tuse_df["TotalCharges"], c=tuse_df["cluster"])
axes.set_title("ThePhoneSubscribers")
axes.set_xlabel("MonthlyCharges")
axes.set_ylabel("TotalCharges")
axes.grid(which='major',
        color = 'k', 
        linewidth = 0.1)
```

Построим box plot общей внесённой суммы оплаты от кластера.

```python
sb.boxplot(x="cluster", y="TotalCharges", data=tuse_df)
```

Таким образом, мы можем видеть, что на первый взгляд, похоже, что кластеры разделили абонентов ровно на 3 категории по тому, сколько они заплатили за свою связь в сумме. И у нас представители кластера-0 заплатили больше всего, клатера-2 средне, а кластера-1 мало.

Теперь построим сводную таблицу кластер-мигрант. Где 0 - не мигрант; 1 - мигрант (перешёл от другого оператора).

```python
tuse_df.groupby(['Churn', 'cluster'])['Churn'].count().unstack()
```

Построим box plot зависимости суммы платежа в месяц от типа клиентов (мигранты\постояльцы)

```python
sb.boxplot(x="Churn", y="MonthlyCharges", data=tuse_df)
```

Отсюда можем сделать вывод, что подавляющее число клиентов, мигрировавших от другого оператора выбирают средний по величине платёж и среди них мало тех, кто платит много или мало. А в категории большинства среди клиентов-постояльцев, наблюдается помимо любителей среднего платежа, дополнительно отдельное количество тех, кто предпочитает маленький платёж, при этом, на случай высокого платежа, там едва меньше чем для мигрантов.

Построим box plot месячной суммы оплаты от кластера.

```python
sb.boxplot(x="cluster", y="MonthlyCharges", data=tuse_df)
```

Визуально, разбиение на кластеры произошло в первую очерещь по совокупному платежу абонентов. Попробуем сделать выводы, повлиявшие на вклад других факторо в кластеризацию.

Абонентов, мигрировавших с оператора на оператор меньше, чем абонентов постояльцев почти в 3 раза и их подавляющие предпочтения по платежам - средний платёж.

Самый немногочисленный кластер, это кластер-0. В нём у нас находится набор с большинством плательщиков большого платежа, немного плательщиков среднего и совсем нет плательщиков маленького платежа, в кластере-0 наибольшая разница по количеству постояльцев и мигрантов.

В кластере-1 у нас подавляющее большинство абонентов предпочитают платёж на уровне 50 у.е., при этом, это самый большой кластер из 3х, и его добрая половина от большего количества содержит плетельщиков малого платежа (условно, это те, кто ниже 50 у.е.), а это могут быть только представители постояльцев, так как по графикам выше, доля мегрантов во-первых и так невелика, а во-вторых, среди них большинство предпочитают платёж от 50 у.е.. При этом, доля мигрантов в кластере-1 наивысшая среди всех кластеров, поэтмоу, скорее всего они занимают в ней ту часть, где платёже выше 50 у.е..

Кластер-2 наиболее средний по величине оплаты. Там содержатся в большем количестве представители постояльцев и почти в 3 раза меньше мигрантов, которые скорее всего занимают там позиции тех из них, кто платит от 50 до 90.

Таким образом, кластеризация на 3 кластера показала нам разделение граждан-абонентов не только согласно тому, сколько они заплатили в общем, но и согласно тому, какой размер платежа они предпочитают и какой нет (например, представители кластера-0 никогда не платят мало), а также, было выявлено небольшое воздействие на разделение фактора того, является ли абонент мигрированным с другого оператора связи.

Так как заметное большинство плательщиков большого платежа, это клиенты постоянные, есть смысл ряду операторов предложить им более выгодный средний тариф, чтобы они перешли к ним и пополнили средний или более высокий сегменты за слегка более низкую плату, чем на момент их постоянства.
